```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head(3)
```

```python
sns.countplot(data=titanic,x='embarked',hue='sex')
plt.title('Passenger Count by Embarked Port, Separated by Sex')
plt.xlabel('Embarked port')
plt.ylabel('No of passengers')
plt.show()
```

```python
sns.swarmplot(data=titanic, x='pclass', y='age', hue='sex')
plt.title('Age Distribution Across Passenger Classes, Categorized by Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.show()
```

```python
gender_counts = titanic['sex'].value_counts()

labels = gender_counts.index
sizes = gender_counts.values
explode = (0.1, 0)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, colors=['skyblue', 'lightcoral'])
```

```python
correlation_data = titanic[['age', 'fare', 'pclass', 'sibsp', 'parch', 'survived']]
correlation_matrix = correlation_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Selected Titanic Features')
plt.show()
```

```python
sns.violinplot(data=titanic, x='pclass', y='fare', hue='sex', split=True)
plt.figure(figsize=(10, 7))
plt.show()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', palette='viridis', alpha=0.7)
plt.title('Age vs. Fare, Colored by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
